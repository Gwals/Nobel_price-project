The Nobel Prize has been among the most prestigious international awards since 1901. Each year, awards are bestowed in chemistry, literature, physics, physiology or medicine, economics, and peace. In addition to the honor, prestige, and substantial prize money, the recipient also gets a gold medal with an image of Alfred Nobel (1833 - 1896), who established the prize.

![](Nobel_Prize.png)

The Nobel Foundation has made a dataset available of all prize winners from the outset of the awards from 1901 to 2023. The dataset used in this project is from the Nobel Prize API and is available in the `nobel.csv` file in the `data` folder.

In this project, you'll get a chance to explore and answer several questions related to this prizewinning data. And we encourage you then to explore further questions that you're interested in!

```python
# Loading in required libraries
import pandas as pd
import seaborn as sns
import numpy as np

# Start coding here!
# most commonly awarded gender and birth country
nobel_data=pd.read_csv('data/nobel.csv')
#print(nobel_data.info())
top_gender = nobel_data['sex'].mode()[0]
print(top_gender)
top_country = nobel_data['birth_country'].mode()[0]
print(top_country)

# decade that have the highest ratio of US-born Nobel Prize winners to total winners in all categories
nobel_data['decade'] = (nobel_data['year'] // 10) * 10
us_winners = nobel_data[nobel_data['birth_country'] == 'United States of America']
us_winner_counts = us_winners.groupby('decade').size()
total_winner_counts = nobel_data.groupby('decade').size()
ratios = us_winner_counts / total_winner_counts
max_decade_usa = ratios.idxmax()
print(max_decade_usa)

# decade and Nobel Prize category combination had the highest proportion of female laureates
female_data = nobel_data[nobel_data['sex'] == 'Female']
total_counts = nobel_data.groupby(['decade', 'category']).size()
female_counts = female_data.groupby(['decade', 'category']).size()
proportion_female = female_counts / total_counts
max_proportion = proportion_female.idxmax()
max_decade, max_category = max_proportion
max_female_dict = {max_decade: max_category}
print(max_female_dict)

# the first woman to receive a Nobel Prize, and in what category
females_only=nobel_data[nobel_data['sex']=='Female']
first_nobel_year=female_data.sort_values(by='year')
first_woman_info= first_nobel_year.iloc[0]
first_woman_name=first_woman_info['full_name']
print(first_woman_name)
first_woman_category=first_woman_info['category']
print(first_woman_category)

# individuals or organizations have won more than one Nobel Prize throughout the years
price_count=nobel_data.groupby('full_name').size()
repeat_winners=price_count[price_count>1]
repeat_list= repeat_winners.index.tolist()
print(repeat_list)
```
